```python
import pickle
import pandas as pd

with open("player_data.pkl", 'rb') as picklefile:
   player_data = pickle.load(picklefile)
```

```python
player_data.head()
```

```python
player_data = player_data = player_data.ix[:,0:3]
player_data['first'], player_data['last'] = player_data['name'].str.split(' ', 1).str
player_data.head()
```

```python
player_data['letter'] = player_data['last'].astype(str).str[0]
player_data['letter'] = player_data.letter.str.lower()
```

```python
firsts = player_data.sort_values(['name','season'])
firsts.head()
```

```python
with open('firsts.pkl', 'wb') as picklefile:
   pickle.dump(firsts, picklefile)
```
