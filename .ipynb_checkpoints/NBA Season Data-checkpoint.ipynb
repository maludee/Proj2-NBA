```python
# Imports and display options
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from sklearn import preprocessing
from sklearn.cross_validation import cross_val_score
from sklearn.linear_model import LinearRegression
from sklearn.cross_validation import train_test_split
from sklearn.cross_validation import KFold

import seaborn as sns
from seaborn import plt

import statsmodels.api as sm
import statsmodels.formula.api as smf
from statsmodels.formula.api import ols

import pickle

pd.options.display.float_format = '{:20,.2f}'.format
pd.options.display.max_columns = 20
pd.options.display.max_rows = 20


%matplotlib inline
```

# Import & clean data

```python
# Read in the 3 player-level csv files
player_totals = pd.read_csv('data/playertotals.csv', 
                            names = ['name','team','season','minutes_played',
                                    'games_played','games_started','points','field_goals', 'wins',
                                     'losses', 'fg_attempts','fg_pct','fg3_pct','steals'])
salaries = pd.read_csv('data/salaries.csv', names = ['name', 'team', 'season', 'salary'])
advanced_data = pd.read_csv('data/advanced.csv', names = ['name','team','season','age','per'])
```

```python
# Read in the 2 team-level csv files
team_stats = pd.read_csv('data/teamstats.csv',
                        names = ['team','season','games', 'tm_3_pct','tm_fg_pct', 'tm_fg_att',
                                'opp_3_pct', 'opp_fg_pct','opp_fg_att'])
team_pace = pd.read_csv('data/teammisc.csv',
                       names = ['team','season','pace'])
```

```python
# Merge the 2 team-level files
team_data = pd.merge(team_stats, team_pace, on=['team','season'])
```

```python
# Merge the 3 player dataframes into 1 big player dataframe
player_data = pd.merge(player_totals, salaries, on=['name','team','season'])
player_data = pd.merge(player_data, advanced_data, on = ['name', 'team','season'])

# Also add the number of games by team 
player_data = pd.merge(player_data, team_stats[['team','season','games']], on=['team','season'])
```

```python
# Determine who the starting 5 are for each team

# calculate percent started, sort, and count
player_data['pct_started'] = player_data['games_started']/player_data['games']
player_data = player_data.sort_values(['team','season','pct_started'], ascending = False)
player_data['count'] = player_data.groupby(['team','season']).cumcount()+1


# drop any player with a count over 5
player_data = player_data.loc[player_data['count'] <= 5, ]

# calculate win/loss ratio
player_data['win_loss_ratio'] = player_data['wins']/player_data['losses']
```

```python
# Convert salary into a float object 
player_data['salary'] = player_data['salary'].astype('str')
player_data['salary'] = player_data['salary'].str.replace("$","")
player_data['salary'] = player_data['salary'].str.replace(",","")
player_data['salary'] = player_data.salary.astype(float).fillna(0.0)
```

```python
# Get the means by team/season for various features
team_df = player_data.groupby(['team','season'], as_index = False)['points','field_goals','fg_attempts',
                                                                 'fg_pct','fg3_pct','steals', 'salary',
                                                                  'age','per','pct_started','win_loss_ratio'].mean()
```

```python
# Rename the columns so that it's clear they're all averages
team_df.columns = ['team','season','avg_points','avg_field_goals','avg_fg_att',
                   'avg_fg_pct','avg_fg3_pct','avg_steals', 'avg_salary', 'avg_age','avg_per',
                   'avg_pct_started', 'avg_wl_ratio']
```

```python
# Merge averages with the player_data dataframe so that all columns are included
player_data = pd.merge(player_data, team_df[['team','season','avg_points','avg_field_goals',
                                             'avg_fg_att','avg_fg3_pct','avg_steals', 'avg_salary',
                                             'avg_age','avg_per','avg_pct_started', 'avg_wl_ratio']], 
                       on=['team','season'])
player_data.head(6)
```

```python
# Pickle this for use in the other dataframe (first game data)
# with open('player_data.pkl', 'wb') as picklefile:
#    pickle.dump(player_data, picklefile)
```

```python
# Make superstar player function
def link_strength(df, points, salary, per):
    pts_diff = df[points] / df['avg_' + points] - 1
    sl_diff = df[salary] / df['avg_' + salary] - 1
    per_diff = df[per] / df['avg_' + per] - 1
    return ((pts_diff + sl_diff + per_diff) / 3)
#     return sl_diff
```

```python
# Apply to get each player's link strength - how far percentage they are away from the average for their own team
player_data['link_strength'] = link_strength(player_data, 'points', 'salary', 'per')
```

```python
# Sort player data by link strength and season to see who were star players in 2016
player_data.sort_values(['season','link_strength'], ascending = False);
```

```python
# Plot distribution of PER
per_mean = str(round(player_data['avg_per'].mean(), 1))
print(per_mean)

with plt.style.context('fivethirtyeight'):
    plt.hist(player_data['per'], bins = 50)
    plt.xlabel('Player Efficiency Rating')
    plt.ylabel('Count')
    plt.title('Distribution of PER (Mean = ' + per_mean + ')')
    plt.show()
```

```python
# Add a strong link and weak link column with the max and min link strengths, respectively
player_data['strong_link'] = player_data.groupby(['team','season'])['link_strength'].transform('max')
player_data['weak_link'] = player_data.groupby(['team','season'])['link_strength'].transform('min')
```

```python
# Collapse the the table to the team level by averaging everything
team_stats = player_data.groupby(['team','season'], as_index = False).mean()
```

```python
# Drop duplicate columns and unnecessary columns 
def drop_columns(df, column_list):
    for l in column_list:
        df.drop(l, axis = 1, inplace = True)
        
drop_columns(team_stats, ['minutes_played','games_played','games_started','points','field_goals',
                          'fg_attempts','fg_pct','fg3_pct','steals', 'pct_started',
                          'salary','age','per','count','win_loss_ratio','link_strength','wins','losses'])
```

```python
# Merge the team_stats dataframe with the team_data dataframe that was scraped from the web
team_stats = pd.merge(team_stats, team_data[['team','season', 'opp_3_pct',
                                             'opp_fg_pct','opp_fg_att','pace']], on=['team','season'])

drop_columns(team_stats, ['games','avg_points'])
```

```python
# Take the spaces out of the team names
team_stats['team'] = team_stats['team'].str.replace(" ","_")
team_stats['team'] = team_stats['team'].str.replace("/","_")
# Print the final dataframe before it will be used for regression
team_stats.head()
```

```python
# Save this for use in the cross validation process
with open('team_stats.pkl', 'wb') as picklefile:
   pickle.dump(team_stats, picklefile)
```

```python
# Normalize the team_stats dataframe between 0 and 1 (between -1 and 1 would be team_stats.mean()/ same as below)
team_norm = (team_stats - team_stats.min()) / (team_stats.max() - team_stats.min())
#team_norm = (team_stats - team_stats.mean()) / (team_stats.std())
team_norm.head(3)
```

```python
team_norm.shape
```

```python
# # Move the dependent variable (avg_wl_ratio) to the first column for simplicity

# cols = list(team_norm)
# cols.insert(0, cols.pop(cols.index('avg_wl_ratio')))
# team_norm = team_norm.ix[:, cols]
# team_norm.head(3)

cols = list(team_norm)
cols.insert(0, cols.pop(cols.index('log_avg_wl')))
team_norm = team_norm.ix[:, cols]
team_norm.head(3)
```

```python
# View the correlations
team_norm.corr()
```

```python
# Plot most of the variable-to-variable relations as scatterplots
sns.pairplot(team_norm, vars = ['log_avg_wl','avg_field_goals','avg_fg_att',
                                'avg_fg3_pct','avg_steals', 'avg_salary', 'avg_per', 'avg_pct_started',
                               'avg_age','strong_link','weak_link','opp_3_pct','opp_fg_pct', 'opp_fg_att',
                                'pace'], size = 1.2, aspect=1.5);
```

```python
# Function to make the r_formula based on the column names in the dataframe
def create_r_formula(dataframe):
    Y = list(dataframe)[0]
    x = ""
    lst = list(dataframe)
    for item in lst[1:]:
        if x is "":
            x = item
        else:
            x = x + " + " + item
    R_form = Y + " ~ " + x
    return R_form
  
```

```python
lm1 = smf.ols(create_r_formula(team_norm),
              data=team_norm)
# Fit the model
fit1 = lm1.fit()
# Print summary statistics of the model's performance
fit1.summary()
```
